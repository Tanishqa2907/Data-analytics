```python
df1 = pd.DataFrame({
    "Name": ["John", "Jane", "Alice", "Bob"],
    "City": ["New York", "Los Angeles", "Chicago", "Houston"],
    "Age": [30, 25, 28, 35]
})

df2 = pd.DataFrame({
    "City": ["New York", "Los Angeles", "Chicago", "Houston"],
    "Age": [30, 25, 28, 40],
    "Population": [8000000, 4000000, 2700000, 2300000]
})

# Merge on multiple columns
merged_multi_df = pd.merge(df1, df2, on=["City", "Age"], how="inner")
print(merged_multi_df)
```

```python
df1 = pd.DataFrame({
    "Country": ["USA", "Canada", "India", "Germany"],
    "Capital": ["Washington", "Ottawa", "New Delhi", "Berlin"]
})

df2 = pd.DataFrame({
    "Country": ["USA", "India", "Germany", "France"],
    "Currency": ["USD", "INR", "Euro", "Euro"]
})

# Inner Join
inner_join_df = pd.merge(df1, df2, on="Country", how="inner")
print(inner_join_df)
```

```python
#Merge two DataFrames and keep all rows (full outer join)
full_outer_merge_df = pd.merge(df1, df2, on="Country", how="outer")
print(full_outer_merge_df)
```

```python
# 1. Creating and Saving Student Data (CSV & Excel)
student_data = pd.DataFrame({
    'Student ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Grade': ['A', 'B', 'C', 'B', 'A'],
    'Score': [85, 78, 45, 90, 30]
})
```

```python
# Save to CSV and Excel
student_data.to_csv("student_data.csv", index=False)
student_data.to_excel("student_data.xlsx", index=False)
```

```python
# 2. Creating and Saving City Data (CSV & Excel)
city_data = pd.DataFrame({
    'City': ['New York', 'Tokyo', 'London', 'Paris', 'Sydney'],
    'Country': ['USA', 'Japan', 'UK', 'France', 'Australia'],
    'Population': [8419600, 13929286, 8982000, 2148000, 5312000],
    'Area': [789, 2194, 1572, 105, 1687]  # Area in square km
})
```

```python
# Save to CSV and Excel
city_data.to_csv("city_data.csv", index=False)
city_data.to_excel("city_data.xlsx", index=False)
```

```python
# 3. Uploading and Modifying Student Data
student_df = pd.read_csv("student_data.csv")
student_df['Pass/Fail'] = student_df['Score'].apply(lambda x: 'Pass' if x > 50 else 'Fail')
student_df.to_csv("student_data_modified.csv", index=False)
student_df.to_excel("student_data_modified.xlsx", index=False)
```

```python
# 4. Uploading and Modifying City Data
city_df = pd.read_csv("city_data.csv")
city_df['Population Density'] = city_df['Population'] / city_df['Area']
city_df.to_csv("city_data_modified.csv", index=False)
city_df.to_excel("city_data_modified.xlsx", index=False)
```
